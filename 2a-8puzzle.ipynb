```python
from copy import deepcopy as dcp
#                     x  y             x  y            x   y            x  y 
Yonler = {"Yukari": [-1, 0], "Asagi": [1, 0], "Sola": [0, -1], "Saga": [0, 1]} 
InitialMatrix=[[7, 2, 4],
               [5, 0, 6],
               [8, 3, 1]]

Goal=      [[0,1, 2], 
            [3, 4, 5], 
            [6, 7, 8]]

class Node:
    def __init__(self, current, previous, x, y, dir):
        (self.current,self.previous,self.x,self.y,self.dir)=(current,previous,x,y,dir)
def position(active_state, e):
    for x in range(len(active_state)):
        if e in active_state[x]:
            return (x, active_state[x].index(e))

def EuclideanCostCalculation(active_state):
    total_cost = 0
    for x in range(len(active_state)):
        for y in range(len(active_state[0])):
            total_cost += abs(x - position(Goal, active_state[x][y])[0]) + abs(y - position(Goal, active_state[x][y])[1])
    return total_cost

def findPath(setOfLimitedRange):
    node = setOfLimitedRange[str(Goal)]
    branch = list()
    while(node.dir):
        branch.append({
            'dir': node.dir,
            'node': node.current
        })
        node = setOfLimitedRange[str(node.previous)]
    branch.append({
        'dir': '',
        'node': node.current
    })
    branch.reverse()
    return branch

def adjacents(node):
    adjacent_nodes = []
    for dir in Yonler.keys():
        newPosition = (position(node.current, 0)[0] + Yonler[dir][0], position(node.current, 0)[1] + Yonler[dir][1])
        if 0 <= newPosition[0] < len(node.current) and 0 <= newPosition[1] < len(node.current[0]):
            newState = dcp(node.current)
            newState[position(node.current, 0)[0]][position(node.current, 0)[1]] = node.current[newPosition[0]][newPosition[1]]
            newState[newPosition[0]][newPosition[1]] = 0
            adjacent_nodes.append(Node(newState, node.current, node.x + 1, EuclideanCostCalculation(newState), dir))

    return adjacent_nodes

def Choose(listForChoosing):
    in_state = True
    for i in listForChoosing.values():
        if in_state or (i.x+i.y) < option:
            in_state = False
            choosing = i
            option = choosing.x+choosing.y
    return choosing

def eightPuzzle(puzzle):
    my_puzzle = {str(puzzle): Node(puzzle, puzzle, 0, EuclideanCostCalculation(puzzle), "")}
    second_puzzle = {}
    while 1==1:
        sample = Choose(my_puzzle)
        second_puzzle[str(sample.current)] = sample
        if sample.current == Goal:
            return findPath(second_puzzle)

        adj_node = adjacents(sample)
        for node in adj_node:
            if str(node.current) in second_puzzle.keys() or str(node.current) in my_puzzle.keys() and ( 
            my_puzzle[str(node.current)].x+my_puzzle[str(node.current)].y )< (node.x+node.y):
                continue
            my_puzzle[str(node.current)] = node

        del my_puzzle[str(sample.current)]

def printMatrix(matrix): 
    global total_steps
    total_steps+=1
    for row in matrix:
        for col in row:
            print(" ", col," ",end="")
        print("")
    print("**************")
    
if __name__ == '__main__':
    initial_matrix = eightPuzzle(InitialMatrix)
    total_steps=0
    for node in initial_matrix:
        printMatrix(node['node'])
        print()
    total_steps= len(initial_matrix) 
    print('Total number steps : ',total_steps-1)


```
